## 7개 지역 보행 노인 사고 위치 지도 시각화

```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd
m = folium.Map(location=[37.511877, 127.059505])
```

```python
df = pd.read_excel('C:/Users/user/Desktop/project_data/7개 지역 피해자 유형 위경도.xlsx')
df = df[["위도", "경도"]]
df.head()
```

```python
get_store_name_loc = df.loc[:,['위도','경도']]
lat =list(get_store_name_loc['위도']) 
lng =list(get_store_name_loc['경도'])
marker_cluster = MarkerCluster().add_to(m)

for i in range(len(lat)):
    folium.Marker(
    location = [lat[i],lng[i]],
    icon=folium.Icon(color='cadetblue',icon='ok'),
    popup = [lat[i], lng[i]]
    ).add_to(marker_cluster)

m.save('C:/Users/user/Desktop/project_data/map1.html')
```
