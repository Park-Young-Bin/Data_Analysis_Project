## 각 지역별 사고다발구역 주변 특징 시각화 코드

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

mpl.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
```

## 1. 경기도 이천 사고 지점 부근 특징

```python
df1 = pd.read_excel('C:/Users/user/Desktop/project_data/9개 지역별 사고지점 주변 특징.xlsx', sheet_name = 0)[['범주']]
```

```python
order_list = df1['범주'].value_counts().index.tolist()
```

```python
plt.figure(figsize = (7, 5))
sns.countplot(data = df1, x = '범주', order = order_list, palette = 'coolwarm')
plt.title('경기도 이천시 사고다발구간 주변 특징', fontsize = 16)
plt.xlabel('주변 시설', fontsize = 13)
plt.ylabel('사고건수', fontsize = 13)
plt.show()
```
