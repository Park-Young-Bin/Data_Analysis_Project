## 지역별 기온 데이터 병합 코드

```python
import pandas as pd
```

## 경기도 이천

```python
df = pd.read_excel('project_data/시간 전처리/시간_전처리_이천.xlsx')
df.head(3)
```

```python
df_temp = pd.read_excel('project_data/지역별 기온 데이터/경기_이천_기온.xlsx')
df_temp.head(3)
```

```python
merge_temp = pd.merge(df, df_temp, how = 'inner', left_on = '사고날짜', right_on='날짜')
merge_temp.head(3)
```

```python
merge_temp.to_excel('project_data/시간 및 기온 전처리 완료/전처리_이천_기온.xlsx', index=False)
```
