## 부족함 경험 여부

```python
label = ['부족함', '부족하지 않음']
size = [129, 73]

plt.figure(figsize = (7,7))
plt.title('횡단중 보행신호시간 부족함 경험 여부', fontsize = 20)

plt.pie(size, 
        startangle = 90, 
        autopct = '%.2f%%', 
        colors = ['orangered', 'darkkhaki'], 
        explode = (0.1, 0), 
        shadow = True, 
        labels = label,
       textprops = {'fontsize':13})

plt.savefig('C:/Users/user/Desktop/project_data/g1.png')
```

## 보행신호시간 체감 정도

```python
df['불편 정도'].value_counts()
```

```python
label = ['매우 적당', '적당', '보통', '짧음', '매우 짧음']
size = [21, 26, 35, 89, 31]

plt.figure(figsize = (7,7))
plt.title('횡단중 보행신호시간 체감 정도', fontsize = 20)

plt.pie(size, 
        startangle = 90, 
        autopct = '%.2f%%', 
        explode = (0,0,0,0.1,0),
        colors = ['papayawhip', 'khaki', 'darkkhaki', 'orangered', 'coral'], 
        shadow = True, 
        labels = label,
       textprops = {'fontsize':13},
       counterclock=False)
plt.savefig('C:/Users/user/Desktop/project_data/g2.png')
```
