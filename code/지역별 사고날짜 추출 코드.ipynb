## 지역별 사고날짜 추출 코드

```python
import pandas as pd
```

```python
df_이천_전처리 = pd.read_excel('project_data/시간, 기온 전처리 전//전처리_이천.xlsx').iloc[:,1:]
df_이천_기온 = pd.read_excel('project_data/지역별 기온 데이터//경기_이천_기온.xlsx')

def func(x):
    result = str(x)
    return result[:4] + '-' + result[4:6] + '-' + result[6:8]
    
df_이천_전처리['사고날짜'] = df_이천_전처리['사고번호'].apply(func)

df_이천_전처리.head(2)
```
